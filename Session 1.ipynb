# logical Operators in Pandas
* `&` and,
* `|` or,
* `~` not

```python
f500[f500["country"] == "India"]
```

```python
f500[(f500["country"] == "USA") & (f500["sector"] == "Technology")]
```

```python
f500[(f500["country"] == "USA") & ~(f500["sector"] == "Technology")]
```
